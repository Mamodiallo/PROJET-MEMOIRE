```python
# 1) Mapper les codes en libellés
dispo_map = {
    1: "Complété",
    2: "Abandonné",
    0: "Interrompu",                # ou "Non démarré" si vous préférez
    6: "Éliminé par présélection",  # si ce code existe
    5: "Quota atteint/fermé"        # idem
}
df["disposition"] = df["Codededisposition"].map(dispo_map).fillna("Autre")

# 2) Compter et calculer les pourcentages
counts = (
    df["disposition"]
      .value_counts(dropna=False)
      .rename_axis("disposition")
      .reset_index(name="count")
)
counts["percent"] = counts["count"] / counts["count"].sum()

# 3) Tracer le camembert
fig = px.pie(
    counts,
    names="disposition",
    values="count",
    hole=0.2,
    title="Répartition des statuts (Codededisposition)"
)
fig.update_traces(
    textinfo="label+percent+value",
    textposition="outside",
    showlegend=False
)
fig.update_layout(margin=dict(t=50, b=0, l=0, r=0))

# 4) Afficher inline dans le notebook
fig.show()
```
